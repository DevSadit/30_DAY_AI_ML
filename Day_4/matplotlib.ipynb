```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
x
```

```python
y
```

```python
# functional method
plt.plot(x, y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x, y)

plt.subplot(1,2,2)
plt.plot(y, x)
```

```python
# object oriented
fig = plt.figure()
fig
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes
axes.plot(x,y)
# axes.set_xlabel("x label")
# axes.set_ylabel("y label")
# axes.set_title("title")
```
