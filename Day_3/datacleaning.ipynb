```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Alice"],
    "Age": [25, 30, 35, np.nan, 29, 25],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "HR"],
    "Salary": [50000, 60000, 70000, 62000, np.nan, 50000]
}
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.dropna(how= "any") # any => remove the whole row if there is one null value
```

```python
df.dropna(how="all") # all => remove only the row where all the value is null
```

```python
# fillna => to fill the value
df["Age"].fillna(df["Age"].mean())
```

```python
df["Salary"].fillna(df["Salary"].median())
```

```python
# forward fill
```

```python
# backward fill
df["Age"].fillna(method= "bfill")
```

```python
# replace the data 
df["Name"] = df["Name"].replace("Charlie", "kola")
df["Name"]
```
