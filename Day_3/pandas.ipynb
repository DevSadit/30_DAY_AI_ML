```python
import pandas as pd
```

```python
# series
s = pd.Series([10, 20, 30, 40])
```

```python
# add names
s.name = "Calories"
s
```

```python
# data type
s.dtype
```

```python
# check values
s.values
```

```python
# check index.
s.index
```

```python
# indexing
s[1:3]
```

```python
# iloc => number location based indexing
s.iloc[[0, 2,3]]
# s.iloc[2]
```

```python
index = ["aam", "jam", "katal", "kola"]
s.index = index
s
```

```python
# accesing data with index names
s["aam"]
```

```python
s.iloc[2]
```

```python
# loc -> label based indexing
s.loc["kola"]
```

```python
# in label based indexing our start and end value included
s["jam":"katal"]
```
