```python
import numpy as np
```

```python
# array creating
arr = np.array([12, 34,56,90])
```

```python
# array slicing
arr[1:3]
```

```python
# last two elements
arr[-1: -5: -1]
```

```python
# filtering the data with boolean filtering
arr[arr<50]
```

```python
# basic math
arr+5
arr-1
arr*2
arr/2
```

```python
# aggregate fn
arr.mean()
arr.max()
arr.min()
arr.std()
# arr
```

```python
# array multiplication, addition, substraction
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

mul = np.dot(A, B)
mul
```

```python
# Broadcasting
b_array = np.array([1, 2])
A + b_array
```
