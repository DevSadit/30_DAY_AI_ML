```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating a basic line plot
x = pd.array([1, 2, 3])
y = pd.array([4, 5, 6])
plt.plot(x, y)
plt.show()
```

```python
# scatter plot
plt.scatter(x, y)
```

```python
# Fake sample data
students = ["A","B","C","D","E","F","G","H"]
study_hours = np.array([1,2,3,4,5,6,7,8])
scores = np.array([35, 50, 55, 60, 65, 70, 82, 90])
ages = np.array([15,16,17,15,16,17,16,15])
marks_distribution = np.random.normal(70, 10, 200)
```

```python
# bar chart
plt.figure(figsize=(8, 5))

plt.bar(students, scores, color="purple")

plt.title("Scores of Students")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.grid(axis="y", alpha=.4)

plt.show()
```

```python
# histogram ****
plt.figure(figsize=(7, 5))

plt.hist(marks_distribution, bins=20, color="orange", alpha=0.7)

plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```
