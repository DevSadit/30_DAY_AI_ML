```python
import numpy as np
```

```python
# first three element
arr= np.array([3, 6, 9, 12, 15])
arr[0:3:1]
```

```python
# last two element
arr[-1:-3:-1]
```

```python
# element from index 1 to 4
arr[1:5]
```

```python
# filtering the data with boolean filtering
arr[arr>10]
```

```python
# Basic Math
arr+5
arr*5
arr/5
arr**2
```

```python
# Aggregate Functions
# mean 
arr.mean()
```

```python
# sum 
arr.sum()
```

```python
# min 
arr.min()
```

```python
# max
arr.max()
```

```python
#  standard deviation (std)
arr_2 = np.array([2, 2, 2, 2, 2, 2])
arr_2.std()
```

```python
# reshape
a = np.array([1,2,3,4,5,6])
a.reshape(2,3)
```

```python
# matrix operation
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

A + B # addition
A - B # substraction
A @ B # multiplication
np.dot(A,B)  # multiplication
np.transpose(A) # transpose
```

```python
# Broadcasting
array = np.array([[1, 2, 3], [4, 5, 6]])
v = np.array([10,20,30])

array + v
```

```python
# Axis Practice
arr_3 = np.array([[2, 4, 6], [1, 3, 5]])
np.sum(arr_3, axis=0) # axis 0 indicate sum on all collum
np.sum(arr_3, axis=1) # axis 1 indicate sum on all row
```
