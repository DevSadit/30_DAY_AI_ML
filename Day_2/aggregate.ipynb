```python
import numpy as np
```

```python
# Aggregations: Min, Max, and Everything In Between
narr= np.array([1, 2, 3, 4, 5, 6, 7])
```

```python
# summing all the values in the array
np.sum(narr)
```

```python
# minimum value in an array
np.min(narr)
```

```python
# maximum value in an array
np.max(narr)
```

```python
np.std(narr)   
```
