# Data concatenation
stacking dataframes on top of each other(rows) or side by side(columns)

```python
df1 = pd.DataFrame({
    "id": [1,2,3],
    "name": ["Adaam", "Bob", "Charlie"]
})

df2 = pd.DataFrame({
    "id": [4,5,6],
    "name": ["Rakib", "Ehsan", "Murphy"]
})

df1
```

```python
df2
```

```python
row_concat = pd.concat([df1, df2], ignore_index=True)
row_concat
```

```python
col_concat = pd.concat([df1, df2], axis=1)
col_concat
```
