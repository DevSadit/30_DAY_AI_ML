```python
import pandas as pd
```

```python
df  = pd.read_csv("raw_data.csv")
df
```

Transforming data
1. apply()
2. map()
3. assign()
4. replace(old, new)
5. rename(colums={old: new})
6. sort_values()
7. sort_index()
8. reset_index()
9. rank()
10. Reorder collums

```python
# apply()
df_2 = df.copy()
df_2["tax"] = df_2["income"].apply(lambda x: "20%" if x >= 50000 else "10%" )
# df_2
```

```python
# map()
gender_map = {"Male": "M", "Female": "F", "Unknown": "U"}
df_2["gender"] = df_2["gender"].map(gender_map)
df_2
```

```python
# assign()
df_2 = df_2.assign(new_income = df_2["income"] * 2)
```

```python
df_2
```

```python
# replace (old, new)
df_2["country"] = df_2["country"].replace("USA", "US")
df_2
```

```python
# rename()
# df_2.columns = df_2.columns.str.upper() 

df_2 = df_2.rename(columns={"income": "salary"})
df_2
```

```python
# sort_values()
df_2 = df_2.sort_values("salary")
df_2
```

```python
# rank()
df_2["Ranking"] = df_2["salary"].rank(ascending=False, method="max")
df_2
```
