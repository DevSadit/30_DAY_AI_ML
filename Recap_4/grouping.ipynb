```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
df2 = df.copy()

df2 = df2.drop_duplicates()
df2 = df2.sort_values("income")
df2 = df2.reset_index(drop=True)
df2
```

Grouping & Aggregation of Data
1. groupby()
2. mean/sum/max/min/count etc
3. agg()

```python
# each countries person avg income
df2.groupby("country")["income"].mean()

# avg income by gender
df2.groupby("gender")["income"].mean()


# all data(min, max and avg) for each country
df2.groupby("country")["income"].agg(["mean", "max", "min"])

# all data(min, max and avg) for each country, with changing collum name.
df2.groupby("country")["income"].agg(avg_salary="mean", maxi_salary="max", min_salary="min")
```
