```python
import pandas as pd
```

Transform data
1. Melt - wide to long format 
2. Piviot - long to wide format

```python
df = pd.DataFrame(
    {
        "country": ["USA", "USA", "BD", "BD"],
        "year": [2020, 2021, 2020, 2021],
        "sales": [100, 120, 90, 110],
        "profit": [20, 25, 18, 22],
        "tax": [20, 12, 20, 18],
    }
)
df
```

```python
df.melt(
    id_vars=["country", "year"],
    value_vars=["sales", "profit", "tax"],
    var_name="matrics",
    value_name="value",
)
```

```python
df2 = pd.DataFrame({
    "Name": ["A", "B"],
    "Math": [85, 78],
    "Physics": [90, 82],
    "Chemistry": [88, 80]
})
df2
```

```python
df2.melt(
    id_vars="Name",
    value_vars=["Math", "Physics", "Chemistry"],
    var_name="Subjects",
    value_name="Marks",
)
```
