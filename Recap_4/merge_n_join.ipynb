```python
import pandas as pd
```

Merging and joining the data
1. Inner Join --> merge( on"common_col_in_tables") 
2. Left Join --> merge(how="left")
3. Right Join --> merge(how="right")
4. Outer Join --> merge(how="outer")

```python
# 1. Customer information table
customers_df = pd.DataFrame(
    {
        "customer_id": [101, 102, 103, 104],
        "name": ["Alice", "Bob", "sadit", "Rana"],
        "city": ["Dhaka", "Chittagong", "Chittagong", "Sylhet"],
    }
)

# 2. Order details table
orders_df = pd.DataFrame(
    {
        "customer_id": [101, 104, 105, 103],
        "order_id": [5001, 5002, 5003, 5000],
        "product": ["Laptop", "Mouse", "Dildo", "Keyboard"],
        "price": [85000, 1500, 3000, 9000],
    }
)

customers_df
```

```python
orders_df
```

```python
inner_merge = pd.merge(customers_df, orders_df, on="customer_id")
inner_merge
```

```python
right_merge = pd.merge(customers_df, orders_df, how="right")
right_merge
```

```python
left_merge = pd.merge(customers_df, orders_df, how="left")
left_merge
```

```python
outer_merge = pd.merge(customers_df, orders_df, how="outer")
outer_merge
```
