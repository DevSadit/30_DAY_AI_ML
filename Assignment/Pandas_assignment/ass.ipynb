```python
import pandas as pd
```

```python
# get the csv
df = pd.read_csv("IRIS.csv")
```

```python
# question 1A
df.head(10)
print(df.shape)
print(df.dtypes)
print(df.describe())
```

```python
# question 1B
# Select only those rows where: 1. petal_length > 4.5, 2. species = "Iris-virginica"
filtered_data = df[(df["petal_length"] > 4.5) & (df["species"] == "Iris-virginica")]
filtered_data
```

```python
# question 1C
# Group by species and compute: Average sepal_length, Maximum petal_width, Standard deviation of sepal_width

grouped_stats = df.groupby("species").agg({
    "sepal_length": "mean",
    "petal_width": "max",
    "sepal_width": "std"
})
print(grouped_stats)
```

```python
# question 1D. 
# Create a new column: "petal_ratio" = petal_length / petal_width
# â€¢ For each species, find the average petal_ratio.
df2 = df.copy()
df2["petal_ratio"] = df["petal_length"] / df["petal_width"]

avg_ratio = df2.groupby("species")["petal_ratio"].mean()
print(avg_ratio)
```
