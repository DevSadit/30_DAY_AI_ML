```python
# selecting collum data
df[["country", "city", "pm25"]]
```

```python
# selecting row data
# 1. loc --> label based accesing data
# 2. iloc --> index based accesing data
df.loc[0:3] # start idx : end idx (inclusive)
df.iloc[0:2] # start idx : end idx (exclusive)
```

```python
# selecting individual cells
df.loc[0:2, ["country", "city", "pm25"]]
df.iloc[0:3, [1,2,5]]
```

```python
# select single scalar value
df.at[0, "city"]
df.iat[0, 3]
```

```python
# view and copy data
cities = df["city"].copy()
cities[0] = "Dhaka"
```

```python
# filtering data
# finding cities which aqi is greater then 100
aqi_temp = df[(df["aqi"] > 100) & (df["temperature"] > 35)][["city", "country"]]
aqi_temp.iloc[0]
# aqi_temp.loc[6]
```

```python
# query --> return a copy
df.query("aqi > 150 & temperature > 30")[["country", "city", "pm25"]]
```

```python
# data cleaning
# -> handle missing values
rd = pd.read_csv("raw_data.csv")
rd.isnull()
rd.isna().sum()
rd.dropna(axis=1)
# rd.fillna(0)
# rdf = rd.ffill()
# rdb = rd.bfill()
cleaned_data = rd.copy()
cleaned_data["age"] = cleaned_data["age"].fillna(cleaned_data["age"].mean())
# cleaned_data
```

```python
# --> handle duplicate values
# 1. dtypes
# 2. astypes(new type)
# 3. to_datetime()

rd["country"].duplicated()
rd.drop_duplicates()
```
