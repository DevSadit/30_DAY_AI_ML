```python
import pandas as pd
import numpy as np
sales = {
    "Product": ["Laptop", "Laptop", "Mouse", "Mouse", "Keyboard"],
    "Price": [800, 850, 20, 25, 40],
    "Quantity": [5, 7, 50, 40, 20],
}
df = pd.DataFrame(sales)
df
```

```python
df.rename(columns={"Price": "Unitprice"})
```

```python
# add total sale collum
df["TotalSale"] = df["Price"] * df["Quantity"]
df
```

```python
# total revenue
total_revenue = df["TotalSale"].sum()
avg_price = df["Price"].mean()
highest_revenue_product = df.loc[df["TotalSale"].idxmax(), "Product"]
highest_revenue_product
```
