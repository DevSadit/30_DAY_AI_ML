```python
def square(x):
    sqr = x*x
    print(sqr)

print('Enter a number:')
x = int(input())
square(x)
```

```python
def even_or_odd(x):
    if(x%2 == 0):
        print("x is even")
    else :
        print("x is odd")

print('Enter a number:')
x = int(input())
even_or_odd(x)
```

```python

def total(lists):
    sum = 0
    for val in lists:
        sum += val
    print(sum)
        
list = [12, 34, 56, 34, 12, 89]

total(list)
```

```python
def reverse(s):
    rev = s[::-1]
    print(rev)

s = "kalo"
reverse(s)
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Example
x = int(input())
print(f"Factorial of {x} is {factorial(x)}")
```
